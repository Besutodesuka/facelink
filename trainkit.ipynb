## **Click to load your data and apply augmentation**


```
# Accessing My Google Drive
from google.colab import drive
drive.mount('/content/drive')
# defining global variable path
# Location of data on Google Drive
image_path =input()
#"/content/drive/My Drive/Colab Notebooks/facelink/imagenet dataset/case3/train/"
```

```
%%capture
from besuto.images import augment 
datagen = augment.imagedata(augment = True, extract = True)
X_train,y_train = datagen.load_faces_dataset(image_path)
```

## **train your Model**


1.   first run this section and type your location that you want model to be
for example '/content/drive/My Drive/modelOutput/'
2.   then type model version (this will be suffig of your model name)




```
import tensorflow as tf
from sklearn.neural_network import MLPClassifier
import pickle
version = input('version of model : ')
model = MLPClassifier(hidden_layer_sizes=(400, 200))
model.fit(X_train,y_train)
filename = '/content/model{}.sav'.format(version)
pickle.dump(model, open(filename, 'wb'))
```
